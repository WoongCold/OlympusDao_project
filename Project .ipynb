```python
total_df_high = pd.DataFrame(columns=df_high['event'].unique())
for i in tqdm(range(len(day_list))):
    action_df = pd.DataFrame(Counter(df_high[df_high['time'] == day_list[i]]['event']).items())

    columns = Counter(df_high[df_high['time'] == day_list[i]]['event']).keys()
    data = Counter(df_high[df_high['time'] == day_list[i]]['event']).values()

    action_df = pd.DataFrame(data=data).T
    action_df.columns = columns

    total_df_high = pd.concat((total_df_high, action_df))
```
